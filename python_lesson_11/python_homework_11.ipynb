```python
from sympy.plotting import plot 
from sympy import *
init_printing()
```

```python
x = Symbol('x', domain=S.Reals)
```

```python
a = 18
b = 5
c = 10
fx = a * x ** 3 + b * x ** 2 + c * x - 30
fx
```

```python
#Определить корни:
solveset(fx, x, Interval(-20, 20)).evalf()
```

```python
#Найти интервалы, на которых функция возрастает и убывает
F=fx.diff(x)
print(F)
solve(F, x)
plot(F, (x, -20, 20))
```

```python
#Построить график:
plot(fx, (x, -20, 20))
```

```python
#Определить промежутки, на котором f > 0
solve_univariate_inequality(fx > 0, x)
```

```python
#Определить промежутки, на котором f < 0
solve_univariate_inequality(fx < 0, x)
```
